```python
import os
import numpy as np
import ffmpeg
from scipy.io import wavfile
from pydub import AudioSegment

input_dir = r"C:\Users\quanz\Documents\UM\Projects\GLX_Project\DNN\MiND_Stimili\Speech"
output_dir = r"C:\Users\quanz\Documents\UM\Projects\GLX_Project\DNN\MiND_Stimili\Speech\Segments"


# Set the segment duration (in milliseconds)
segment_duration = 2000  # 2 seconds

# Loop through all audio files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".mp3") or filename.endswith(".wav"):  # Filter audio files
        file_path = os.path.join(input_dir, filename)

        # Load the audio and convert to mono and 16 kHz
        audio = AudioSegment.from_file(file_path)
        audio = audio.set_channels(1)  # Mono
        audio = audio.set_frame_rate(16000)  # 16 kHz

        # Get the total length of the audio in milliseconds
        total_length = len(audio)

        # Loop to create and export each segment
        for i in range(0, total_length, segment_duration):
            segment = audio[i:i + segment_duration]
            segment_filename = f"{os.path.splitext(filename)[0]}_segment_{i // segment_duration + 1}.wav"
            segment_path = os.path.join(output_dir, segment_filename)

            # Export the segment
            segment.export(segment_path, format="wav")
            print(f"Exported: {segment_filename}")

print("All audio files processed and segmented successfully!")


# input_file = r'C:\Users\quanz\Documents\UM\Projects\GLX_Project\DNN\MiND_Stimili\Test\Violin_mp3.mp3'
# output_file = r'C:\Users\quanz\Documents\UM\Projects\GLX_Project\DNN\MiND_Stimili\Test\output.wav'  

# # Use the full path to the ffmpeg executable
# ffmpeg.input(input_file).output(output_file).run(cmd='C:\\ffmpeg\\bin\\ffmpeg.exe')

# # After conversion, read the output WAV file
# # from scipy.io import wavfile

# sr, audio_signal = wavfile.read(output_file)

# if audio_signal.ndim == 2:  # Check if it's stereo
#     audio_signal = np.mean(audio_signal, axis=1)

# audio_signal = audio_signal.astype(np.int16)

# import matplotlib.pyplot as plt
# sample_rate, audio_signal = wavfile.read("segment_2.wav")

# print("Shape of audio signal:", audio_signal.shape)
# print("Data type of audio signal:", audio_signal.dtype)

```
